```python
import os
import pandas as pd


# Clean whitespace from important columns
df['Pattern'] = df['Pattern'].astype(str).str.strip()
df['Code'] = df['Code'].astype(str).str.strip()

# Base directory of input folders (read-only)
input_base = "//kaggle/input/raw-final/Raw Data "

# Output directory (writable)
output_base = "/kaggle/working/enriched_data"
os.makedirs(output_base, exist_ok=True)

# Loop through each pattern folder
for pattern_folder in os.listdir(input_base):
    pattern_path = os.path.join(input_base, pattern_folder)

    if not os.path.isdir(pattern_path):
        continue

    print(f"Processing folder: {pattern_folder}")

    for filename in os.listdir(pattern_path):
        if filename.endswith(".csv"):
            sample_id = filename.replace(".csv", "").strip()  # e.g., "201"

            # Match metadata row where pattern and sample ID match
            match = df[
                (df['Pattern'] == pattern_folder.strip()) &
                (df['Code'].str[-3:] == sample_id)
            ]

            if match.empty:
                print(f"No match found for {pattern_folder}/{filename}")
                continue

            # Load individual sample CSV
            sample_csv_path = os.path.join(pattern_path, filename)
            sample_df = pd.read_csv(sample_csv_path)

            # Add metadata as new columns
            for col in match.columns:
                sample_df[col] = match.iloc[0][col]

            # Prepare output folder and path
            output_folder = os.path.join(output_base, pattern_folder.strip())
            os.makedirs(output_folder, exist_ok=True)

            output_csv_path = os.path.join(output_folder, filename)
            sample_df.to_csv(output_csv_path, index=False)

            print(f"Saved enriched file: {output_csv_path}")
```

# Deleting the unnecessary columns

```python


# ✅ Set the columns you want to delete
columns_to_delete = ['Time', 'Extension','Load']  # ⬅️ Replace with actual column names

# ✅ Path to enriched data
enriched_base = "/kaggle/working/enriched_data"

# Loop over each pattern folder
for pattern_folder in os.listdir(enriched_base):
    folder_path = os.path.join(enriched_base, pattern_folder)

    if not os.path.isdir(folder_path):
        continue

    for filename in os.listdir(folder_path):
        if filename.endswith(".csv"):
            file_path = os.path.join(folder_path, filename)

            # Load the enriched CSV
            df = pd.read_csv(file_path)

            # ✅ Drop the first row
            df = df.iloc[1:].reset_index(drop=True)

            # ✅ Drop the specified columns if they exist
            df = df.drop(columns=[col for col in columns_to_delete if col in df.columns])

            # ✅ Save back to the same file
            df.to_csv(file_path, index=False)
            print(f"Cleaned: {file_path}")
```
